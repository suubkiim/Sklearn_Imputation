```python
dataset_df_na = dataset_df.copy()
# select only numeric columns to apply the missingness to
cols_list = dataset_df_na.select_dtypes('number').columns.tolist()
```

```python
ratio = 0.05
# randomly insert NA values
for col in dataset_df[cols_list]:
    dataset_df_na.loc[dataset_df.sample(frac=ratio).index, col] = np.nan
```

```python
dataset_df_na
```

```python
## https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html
## sklearn simple imputer
"""
strategystring, default=’mean’
The imputation strategy.

If “mean”, then replace missing values using the mean along each column. Can only be used with numeric data.

If “median”, then replace missing values using the median along each column. Can only be used with numeric data.

If “most_frequent”, then replace missing using the most frequent value along each column. Can be used with strings or numeric data. If there is more than one such value, only the smallest is returned.

If “constant”, then replace missing values with fill_value. Can be used with strings or numeric data.

New in version 0.20: strategy=”constant” for fixed value imputation.

fill_valuestring or numerical value, default=None
When strategy == “constant”, fill_value is used to replace all occurrences of missing_values. If left to the default, fill_value will be 0 when imputing numerical data and “missing_value” for strings or object data types.

verboseinteger, default=0
"""
from sklearn.impute import SimpleImputer

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(dataset_df_na)
df_simple_imputed = pd.DataFrame(imp_mean.transform(dataset_df_na))
```

```python
df_simple_imputed
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imputed=imp_mean.fit_transform(dataset_df_na)
test_simple_imputed = pd.DataFrame(imputed)
test_simple_imputed
```

```python
## KNN
from sklearn.impute import KNNImputer
n_neighbors = 5
imp_KNN = KNNImputer(n_neighbors=n_neighbors)
imp_KNN = imp_KNN.fit_transform(dataset_df_na)
```

```python
df_KNN_imputed = pd.DataFrame(imp_KNN, columns=dataset_df_na.columns)
```
